```python
# Task 2.1
myDataFrame = pd.read_csv("student-math.csv", sep=";")
myDataFrame.head()
```

```python
# Task 2.2
myDataFrame['final_grade'] = myDataFrame['G1'] + myDataFrame['G2'] + myDataFrame['G3']
myDataFrame.head()
```

```python
# Task 2.3
myDataFrame.drop(["G1","G2","G3"], axis=1, inplace=True)
myDataFrame.head()
```

```python
# task 2.4
myDataFrame.replace(['yes','no'], [1,0])
```

```python
# task 2.5

plt.title("Studytime vs. FinalGrade")
plt.xlabel("Studytime")
plt.ylabel("FinalGrade")
x = myDataFrame['studytime']
y = myDataFrame['final_grade']

colors = np.where(x==1,'k',
                  np.where(x==2,'b',
                           np.where(x==3,'y',
                                    np.where(x==4,'g','r'))))

plt.scatter(x, y, c = colors)

# task 2.6
myDataFrame.boxplot(by = 'studytime', column = ['final_grade'], grid = False)
```
